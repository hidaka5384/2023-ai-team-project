### yolov4_vegetable.cfg(ネットワークの設定)の作成


##### cfgファイルの変更点

```
[net]
# 2-5行目
# GPUのメモリーオーバーが発生した場合，subdivisionsを2倍にしたり、widthとheightを小さくする
batch=64
subdivisions=64
width=512
height=512

# 19行目
# max_batchesは最大の学習回数(クラス数*2000)
max_batches=6000

# 21行目
# steps(max_batches*0.8, max_batches*0.9)
steps=4800,5400

# [yolo]の1つ手前の[convolutional](3箇所)
[convolutional]
# 961，1049，1137行目付近
# filters=(class数+5)*3=(3+5)*3=24
filters=24

# [yolo](3箇所)
# 968，1056，1144行目付近
classes=3
```
